## <font color ="darkred">Normal Distribution</font>

```python
# example
import numpy as np
ages = [31,30,28,33,22,40,29,31,33,32.5]
print(np.mean(ages))
import matplotlib.pyplot as plt
plt.hist(ages)

#1 tells you if the data has any outliers
#2. helps makes decision under uncertainty
```

```python
plt.boxplot(ages)
```

```python
kijabe = [16,23,15,18,18,17.5,16,12,15,20,19,19,18]
plt.hist(kijabe)
```

```python
mu, sigma = 31,3 # mean and standard deviation
s = np.random.normal(mu, sigma, 1000)
plt.hist(s)
```

```python
import seaborn as sns

sns.distplot(s,hist =True)
```

```python
from ipywidgets import interact,fixed
```

```python
def plotta(x,title,color ="cyan",width =10,height=5):
    plt.figure(figsize = (width,height))
    plt.plot(x,color =color)
    plt.title(title)
    return plt.show()
```

```python
plotta(s,"","Darkorange",10,4)
```

```python
interact(plotta,x = fixed(s),title = "",color ="cyan",width =10,height =5)
```
