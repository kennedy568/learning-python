## In Pandas Dataframe we can iterate an element in two ways:

# Iterating over rows
# Iterating over columns

```python
import pandas as pd

dict = {'name':["aparna", "pankaj", "sudhir", "Geeku"], 
        'degree': ["MBA", "BCA", "M.Tech", "MBA"], 
        'score':[90, 40, 80, 98]}
df = pd.DataFrame(dict)
```

```python
df
```

```python
df.T
```

```python
import pandas as pd
for i , j in df.iterrows():
    print(i, j)
```

```python
for key, value in df.iteritems(): 
    print(key, value) 
    print() 
```

```python
import pandas as pd

data = pd.read_csv('nba.csv')

data.head()
```

```python
for i , j in data.iterrows():
    print(i, j)
    print()
```

```python
for key, value in data.iteritems():
    print(key, value)
    print()
```

## Iterating over Columns :


```python
import pandas as pd 

dict = {'name':["aparna", "pankaj", "sudhir", "Geeku"], 
        'degree': ["MBA", "BCA", "M.Tech", "MBA"], 
        'score':[90, 40, 80, 98]} 

df = pd.DataFrame(dict)

df
```

```python
columns = list(df)

print(columns)


for i in columns:
    print(df[i][2])
```
