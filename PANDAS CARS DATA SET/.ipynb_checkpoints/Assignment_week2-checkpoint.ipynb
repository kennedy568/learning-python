# Bonus Round

* I have made Three types of graphs that take in more data using various libraries available in Python (do not run the cells)
* This is just to show you that you are not limited to Matplotlib or seaborn library and can learn lots of other libraries
* The one below in particular has features that allow you to hover on the graph and see additional details

```python
from plotnine import *
new = df.iloc[0:5,:].melt(id_vars = "Vehicle_types")

(ggplot(new[new.variable.isin(["mpg","disp","qsec","carb"])],aes(x = "Vehicle_types",y = "value",fill = "variable"))+
geom_bar(stat ="identity",position = "dodge")+coord_flip()+
labs(title = "Comparison between four variables in 5 top vehicle types(Bar Plots)"))
```

```python
from plotnine import *
new = df.iloc[0:5,:].melt(id_vars = "Vehicle_types")

(ggplot(new[new.variable.isin(["mpg","disp","qsec","carb"])],aes(x = "Vehicle_types",y = "value",fill = "variable"))+
geom_bar(stat ="identity",position = "fill")+coord_flip()+
labs(title = "Comparison between four variables in 5 top vehicle types (Stacked Plot)"))
```

```python
import plotly.express as px

px.violin(new,x = "Vehicle_types",y = "value",title = "Spread of indicators for various car types(Violin plots)")
# note that violin plots are same as box plots but have a large stomack where most data lie
```
